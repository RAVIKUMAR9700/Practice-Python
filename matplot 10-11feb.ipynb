```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(10,110,10)
y=np.arange(20,220,20)
y1=np.arange(30,330,30)
plt.plot(x,y,marker='s',color='#c70039',mec='blue',markersize=10,linewidth=2,ls=':',label='2020')
plt.plot(x,y1,marker='o',color='#848CE7',mec='m',markersize=10,linewidth=2,ls='-',label='2010')
plt.xlabel('number of item')
plt.ylabel('price of item')
plt.title('Item- price graph')
plt.legend(loc='best',ncol=2,fontsize=12)
plt.show()
```

```python
d={'2005':[100,200,400,500],'2010':[100,500,600,700],'2020':[200,600,700,800]}
df_data=pd.DataFrame(d,index=['bmw','mercedes','audi','tata'])
df_data
```

```python
plt.bar(['bmw','mercedes','audi','tata'],[100,500,600,700])
plt.xlabel('companies name')
plt.ylabel('sales in 2010')
plt.title('company car sales data 2010')
plt.show()
```

```python
cases=[2050,977,943,889,843]
c=['red','orange','yellow','#88ff99','green']
statenames=['maharastra','kerala','karnataka','andra pradesh','tamil nadu']
plt.figure(figsize=(20,5))
plt.barh(statenames,cases,color=c)
plt.xlabel('cases in thousands')
plt.ylabel('state names')
plt.xticks(cases)
plt.title('statewise covid cases')
plt.show()
```
